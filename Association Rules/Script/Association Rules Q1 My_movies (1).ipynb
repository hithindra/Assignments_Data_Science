```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
```

```python
movies = pd.read_csv("C:\ExcelR\Data Scientist\Association Rules\my_movies.csv")
movies.head(10)
```

```python
movies.shape
```

```python
movies2=movies.iloc[:,5:]
movies2
```

# Support =0.1

```python
frequent_itemsets = apriori(movies2, min_support=0.1, use_colnames=True)
frequent_itemsets
```

```python
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.7)
rules
```

```python
import matplotlib.pyplot as plt
plt.scatter(x='support',y='confidence',data=rules)
```

```python
rules.sort_values('confidence',ascending = False)
```

```python
rules[rules.lift<5]
```
